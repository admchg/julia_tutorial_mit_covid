# Download and import

```julia
url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
```

```julia
download(url, "covid_data.csv")
```

```julia
using CSV
using DataFrames
```

```julia
df = CSV.read("covid_data.csv", DataFrame);
```

```julia
rename!(df, 1 => "province", 2 => "country");
```

# Exercise 1

### Add country case counts

```julia
all_countries = unique(collect(df[:, 2]));
```

```julia
countries = rand(all_countries, 10)
```

```julia
num_days = size(df)[2] - 4
```

```julia
country_cases = Dict()
```

```julia
for c in countries
    country_cases[c] = sum(convert(Array, df[df[:, 2] .== c, 5:end]), dims = 1)
end
```

```julia
country_cases
```

### Add dates

```julia
using Dates
```

```julia
col_names = String.(names(df))[5:end];
fmt = Dates.DateFormat("m/d/Y")
dates = parse.(Date, col_names, fmt) .+ Year(2000);
```

```julia
using Plots
```

```julia
p = plot()

for c in countries
    plot!(dates, vec(country_cases[c]),
            label = c)
end
```

```julia
p
```

#### Log-scale

```julia
log_country_cases = Dict()
```

```julia
for c in countries
    log_country_cases[c] = log.(country_cases[c])
    replace!(log_country_cases[c], -Inf=>NaN)
end
```

```julia
p2 = plot()

for c in countries
    plot!(dates, vec(log_country_cases[c]),
            label = c)
end
```

```julia
p2
```

### Interactive plot

```julia
using Interact
```

```julia
@manipulate for i in num_days
    p2 = plot(xlim = (0, num_days))

    for c in countries
        plot!(1:i, vec(log_country_cases[c][1:i]),
                label = c)
    end
    
    p2
end
```

# Exercise 2

### Combine across countries

```julia
df_g = groupby(df, ["country"]);
```

```julia
total_cases_to_date = combine(df_g, names(df_g)[5:end] .=> sum)
```

```julia
cum_cases = convert(Array, total_cases_to_date[:, 9:end]);
cum_cases_minus7 = convert(Array, total_cases_to_date[:, 2:(end - 7)]);

new_cases = cum_cases - cum_cases_minus7
```

```julia
df_new_cases = convert(DataFrame, new_cases);
rename!(df_new_cases, names(total_cases_to_date)[9:end]);
```

```julia
total_cases_to_date = total_cases_to_date[:, 9:end];
```

```julia
replace!(eachcol(df_new_cases), 0 .=> "NaN")
```

```julia
@manipulate for i in 1:(num_days - 7)
    p3 = plot()

    cal_date = names(df)[(11 + i)]
    
    scatter!(total_cases_to_date[:, i],
                    df_new_cases[:, i])
    
    for j in 1:192
        annotate!(total_cases_to_date[j, i],
            df_new_cases[j, i], text(all_countries[j], 10, :black))
    end
    
    p3
end
```
